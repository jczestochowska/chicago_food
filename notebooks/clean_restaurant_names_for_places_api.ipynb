```python
import pandas as pd
```

```python
DATASET_FILEPATH = "/home/jczestochowska/workspace/ada/chicago_food/data/food-inspections.csv"

inspections = pd.read_csv(DATASET_FILEPATH)
```

```python
inspections = inspections.dropna(subset=["Facility Type"])
```

```python
inspections = inspections[inspections["Facility Type"].str.contains("restaurant", case=False)]
```

```python
query_params = inspections[["DBA Name", "Latitude", "Longitude", "Address", "Zip"]]
query_params['DBA Name'] = query_params['DBA Name'].apply(str.lower)
query_params['Address'] = query_params['Address'].apply(str.lower)
```

```python
query_params = query_params.drop_duplicates("DBA Name")
```
